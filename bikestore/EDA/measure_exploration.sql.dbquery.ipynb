```sql
/* MEASURES EXPLORATION */
-- Find the total sales
select 
    sum(sales_amount) as total_sales
from bike_store.gold.fact_sales;

-- Find how many items are sold
select 
    sum(quantity) as total_items
from bike_store.gold.fact_sales;

-- Find the average selling price
select 
    round(avg(price),2) as avg_sales
from bike_store.gold.fact_sales;

-- Find the total number of orders
select 
    count(distinct order_number) as total_orders
from bike_store.gold.fact_sales;

-- Find the total number of products
select 
    count(distinct product_name)
from bike_store.gold.dim_product;

-- Find the total number of customers
select 
    count(customer_key) as total_customers
from bike_store.gold.dim_customer;

-- Find the total number of customer that has placed an order
select 
    count(distinct customer_key) as total_customers
from bike_store.gold.fact_sales;

-- Generate a Report that shows all key metics of the business
select 'Total Sales' as measure_name, sum(sales_amount) as measure_value 
from bike_store.gold.fact_sales
union all
select 'Total Items Sold', sum(quantity) from bike_store.gold.fact_sales
union all
select 'Average Sales', round(avg(price),2) from bike_store.gold.fact_sales
union all
select 'Total Nr. Orders', count(distinct order_number) from bike_store.gold.fact_sales
union all
select 'Total Nr. Products', count(product_name) from bike_store.gold.dim_product
union all
select 'Total Nr. Customers', count(customer_key) from bike_store.gold.dim_customer
union all
select 'Total Customers with Orders', count(distinct customer_key) from bike_store.gold.fact_sales;
```
